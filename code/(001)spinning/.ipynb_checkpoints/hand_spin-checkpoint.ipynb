```python
class ObjectRotate():
    
    states = ['START','END','UNKNOWN']
    
    def __init__(self):
        
        self.cycle_count = 0
        
        self.machine = Machine(model = self, states =ObjectRotate.states, initial = 'UNKNOWN',ignore_invalid_triggers = True)
        
        self.machine.add_transition(trigger = 'unknown', source = 'UNKNOWN', dest = 'UNKNOWN') #1
        self.machine.add_transition(trigger = 'a1', source = 'UNKNOWN', dest = 'A1') #2
        self.machine.add_transition(trigger = 'a2', source = 'UNKNOWN', dest = 'A2') #3
        self.machine.add_transition(trigger = 'unknown', source = 'A1', dest = 'UNKNOWN') #4
        self.machine.add_transition(trigger = 'unknown', source = 'A2', dest = 'UNKNOWN') #5
        self.machine.add_transition(trigger = 'b1', source = 'A1', dest = 'B1') #6
        self.machine.add_transition(trigger = 'b1', source = 'B1', dest = 'B1') #7
        self.machine.add_transition(trigger = 'unknown', source = 'B1', dest = 'B1') #8
        self.machine.add_transition(trigger = 'b2', source = 'A2', dest = 'B2') #9
        self.machine.add_transition(trigger = 'b2', source = 'B2', dest = 'B2') #10
        self.machine.add_transition(trigger = 'unknown', source = 'B2', dest = 'B2') #11
        self.machine.add_transition(trigger = 'c1', source = 'B1', dest = 'C1') #12
        self.machine.add_transition(trigger = 'c1', source = 'C1', dest = 'C1') #13
        self.machine.add_transition(trigger = 'unknown', source = 'C1', dest = 'UNKNOWN') #14
        self.machine.add_transition(trigger = 'c2', source = 'B2', dest = 'C2') #15
        self.machine.add_transition(trigger = 'c2', source = 'C2', dest = 'C2') #16
        self.machine.add_transition(trigger = 'unknown', source = 'C2', dest = 'UNKNOWN') #17
        self.machine.add_transition(trigger = 'a2', source = 'C1', dest = 'D') #18
        self.machine.add_transition(trigger = 'a2', source = 'D', dest = 'UNKNOWN', after = 'update_cycle_count') #19
        self.machine.add_transition(trigger = 'a1', source = 'C2', dest = 'D') #20
        self.machine.add_transition(trigger = 'a1', source = 'D', dest = 'UNKNOWN', after = 'update_cycle_count') #21
        self.machine.add_transition(trigger = 'a1', source = 'B1', dest = 'UNKNOWN') #22
        self.machine.add_transition(trigger = 'a2', source = 'B2', dest = 'UNKNOWN') #23
        
         
    def update_cycle_count(self):
        self.cycle_count += 1
```

```python
def findHands(img, draw=True):
        """
        Finds hands in a BGR image.
        :param img: Image to find the hands in.
        :param draw: Flag to draw the output on the image.
        :return: Image with or without drawings
        """
        results = hand.process(img)
        allHands = []
        h, w, c = img.shape
        if  results.multi_hand_landmarks:
            for handType,handLms in zip(results.multi_handedness,results.multi_hand_landmarks):
                myHand={}
                ## lmList
                mylmList = []
                xList = []
                yList = []
                for id, lm in enumerate(handLms.landmark):
                    px, py = int(lm.x * w), int(lm.y * h)
                    mylmList.append([px, py])
                    xList.append(px)
                    yList.append(py)

                ## bbox
                xmin, xmax = min(xList), max(xList)
                ymin, ymax = min(yList), max(yList)
                boxW, boxH = xmax - xmin, ymax - ymin
                bbox = xmin, ymin, boxW, boxH
                cx, cy = bbox[0] + (bbox[2] // 2), \
                         bbox[1] + (bbox[3] // 2)

                myHand["lmList"] = mylmList
                myHand["bbox"] = bbox
                myHand["center"] =  (cx, cy)

                
                if handType.classification[0].label =="Right":
                    myHand["type"] = "Left"
                else:
                    myHand["type"] = "Right"
                allHands.append(myHand)

                ## draw
                if draw:
                    mp_drawing.draw_landmarks(img, handLms,
                                               mp_hands.HAND_CONNECTIONS)
#                     cv2.rectangle(img, (bbox[0] - 20, bbox[1] - 20),
#                                   (bbox[0] + bbox[2] + 20, bbox[1] + bbox[3] + 20),
#                                   (255, 0, 255), 2)
                    cv2.putText(img,myHand["type"],(bbox[0] - 30, bbox[1] - 30),cv2.FONT_HERSHEY_PLAIN,
                                2,(255, 0, 255),2)

            return allHands
```

```python
def findPosition(allHands, img, handRL = 'Right' , landmarks_no = 0, draw=True):
    coordinate = []
    if allHands:
        for hand in allHands:
            if hand['type'] == handRL:
                x, y = hand['lmList'][landmarks_no][0], hand['lmList'][landmarks_no][1]
                coordinate.append(x)
                coordinate.append(y)
                if draw:
                    cv2.circle(img, (x, y), 5, (0, 255, 0), cv2.FILLED)
                    #cv2.putText(img, str(coordinate),(x, y), cv2.FONT_HERSHEY_DUPLEX, 0.5, (255, 255, 255), 1, cv2.LINE_AA)
    return coordinate
```

```python
def calculateDistance(img, p1, p2):
    length = None
    if len(p1) == 2 and len(p2) == 2 :
        x1, y1 = p1
        x2, y2 = p2
        length = math.hypot(x2 - x1, y2 - y1)
    return length   
```

```python
def calculateMaxDistance(distance_current,distance_max):
    if distance_current:
            if distance_max < distance_current:
                distance_max = distance_current             
    return distance_max
```

```python
def findROI(img, index_finger_pip, max_ift2ifm_distance, draw = True):
    roi_coordinate = []
    if index_finger_pip:
        if max_ift2ifm_distance:
            w = max_ift2ifm_distance 
            h = max_ift2ifm_distance 
            c = [(index_finger_pip[0] + h),(index_finger_pip[1])]
            p1 = [int(c[0]-w),int(c[1]-0.8*h)]
            p2 = [int(c[0]+w/2),int(c[1]+h)]
            roi_coordinate.append(p1)
            roi_coordinate.append(p2)
            #crop_img = img[p1[1]:p2[1], p1[0]:p2[0]]
            if draw:
                #影像, 頂點座標, 對向頂點座標, 顏色, 線條寬度
                cv2.rectangle(img,(p1[0],p1[1]),(p2[0],p2[1]), (255, 0, 255), 2)
    return roi_coordinate
```

```python
def maskImg(binarize_img, lm_list):
    if binarize_img is not None:
        if type(binarize_img) is np.ndarray:
            if binarize_img.shape[0]>1 and binarize_img.shape[1]>1:
                if lm_list:
                    if len(lm_list[0])==2:    
                        lm_y = []
                        for lm in lm_list:
                            lm_y.append(lm[0])                                       
                        lm_max = max(lm_y)+20
                        binarize_img[:,:lm_max] = 0
                        return  binarize_img                       
```

```python
def cropImg(img,roi):
    if img is not None:
        if len(roi) == 2:
            if len(img.shape) == 2 or len(img.shape) == 3:
                crop_img = img[roi[0][1]:roi[1][1], roi[0][0]:roi[1][0]]
                return crop_img
```

```python
def findContour(fg_img, bgr_img):
    h = None
    w = None
    if fg_img is not None:
        if type(fg_img) is np.ndarray:
            if fg_img.shape[0]>1 and fg_img.shape[1]>1:
                resize_ratio = 300/fg_img.shape[0]
                fg_img = cv2.resize(fg_img, (int(fg_img.shape[1]*resize_ratio),(int(fg_img.shape[0]*resize_ratio))), interpolation = cv2. INTER_CUBIC)
                bgr_img = cv2.resize(bgr_img, (int(bgr_img.shape[1]*resize_ratio),(int(bgr_img.shape[0]*resize_ratio))), interpolation = cv2. INTER_CUBIC)
                kernel = cv2.getStructuringElement(cv2.MORPH_RECT, [3,3])
                #fg_img = cv2.morphologyEx(fg_img, cv2.MORPH_OPEN, kernel)
                fg_img = cv2.GaussianBlur(fg_img, (3, 3), 0)
                _,fg_img = cv2.threshold(fg_img,127,255,cv2.THRESH_BINARY)
                contours, _ = cv2.findContours(fg_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
                if len(contours) != 0:
                     # find the biggest countour (c) by the area
                    c = max(contours, key = cv2.contourArea)
                    x,y,w,h = cv2.boundingRect(c)
                    cv2.rectangle(bgr_img,(x,y),(x+w,y+h),(0,255,255),2)
                    cv2.drawContours(bgr_img, contours, -1, (0,255,0), 3)
                                        
    return bgr_img, w, h
```

```python
def objectPose(w,h):
    object_pose = 'unknown'
    object_hight_threshold = 100
    object_area_max_threshold = 15000
    object_area_min_threshold = 1000
    if w:
        if h:
            area = w * h 
            if (h > object_hight_threshold) and (object_area_min_threshold<area<object_area_max_threshold):
                object_pose = '180_deg'
            elif (h < object_hight_threshold) and (object_area_min_threshold<area<object_area_max_threshold):
                object_pose = '0_deg'
    return object_pose
```

```python
def morphology(img):
    if type(img) is np.ndarray:
        if img.shape[0]>1 and img.shape[1]>1:
            resize_ratio = 300/img.shape[0]
            img = cv2.resize(img, (int(img.shape[0]*resize_ratio),int(img.shape[1]*resize_ratio)), interpolation = cv2.INTER_CUBIC)
            
            img = cv2.dilate(img, None, iterations=3)
            img = cv2.erode(img, None, iterations=3)
            return img
```

```python
#video = cv2.VideoCapture(0)
video = cv2.VideoCapture('output.mp4')

# Create background subtractor objects
#backSub_MOG2 = cv2.createBackgroundSubtractorMOG2(history = 5, varThreshold = 32, detectShadows = False)
backSub_KNN = cv2.createBackgroundSubtractorKNN(history = 5, dist2Threshold = 400, detectShadows = False)
max_right_hand_ift2ifm_distance = 0

while True:
    success, img = video.read()
    if not success:
        break
    if success:
        # Update the background model
        fgMask = backSub_KNN.apply(img)
        # Thresholding
        ret,fgMask = cv2.threshold(fgMask,127,255,cv2.THRESH_BINARY)  
        #BGR2RGB
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        # Detection
        allHands = findHands(img, draw=0)
        
        # Find landmark coordinate
        right_index_finger_tip = findPosition(allHands, img, handRL = 'Right' , landmarks_no = 8 , draw=True)  
        right_index_finger_pip = findPosition(allHands, img, handRL = 'Right' , landmarks_no = 6 , draw=True)  
        right_index_finger_mcp = findPosition(allHands, img, handRL = 'Right' , landmarks_no = 5 , draw=True)
        #Find All Right landmark coordinate
        all_right_hand_lm = findHandAllLm(allHands,handRL = 'Right')

        # Find distance between middle_finger_tip & middle_finger_mcp
        right_hand_ift2ifm_distance = calculateDistance(img, right_index_finger_tip, right_index_finger_mcp)
        # Find max distance between middle_finger_tip & middle_finger_mcp
        max_right_hand_ift2ifm_distance = calculateMaxDistance(right_hand_ift2ifm_distance, max_right_hand_ift2ifm_distance)
        # Find ROI
        right_roi = findROI(img, right_index_finger_pip, max_right_hand_ift2ifm_distance, draw=False)
        # RGB to BGR
        img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR) 
        # Mask hand area
        fgMask = maskImg(fgMask, all_right_hand_lm)
        # Crop img roi 
        crop_fg_img = cropImg(fgMask, right_roi)
        crop_color_img = cropImg(img, right_roi)
        img_cnt, w,h = findContour(crop_fg_img, crop_color_img)
        
        #pose = objectPose(w,h)
        #print(pose)
        showVideo(img_cnt, fps = 30)  
        
        # Do morphology
        #crop_color_img = morphology(crop_fg_img)
        # Show img
        #showVideo(crop_fg_img, fps = 30)
        
        if keyboard.is_pressed('esc'):
            break
video.release()
```
